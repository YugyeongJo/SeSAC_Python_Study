### 튜플관련 메서드
- 리스트와 튜플은 모두 시퀀스 타입이지만, 리스트는 가변(mutable)이고, 튜플은 불변(immutable) 가능한 메서드의 차이
- 리스트는 다양한 수정 메서드(append, extend, insert, remove, pop, clear, sort, revers 등)를 가짐
- 튜플은 불변이기 때문에 내용을 수정하는 메서드가 없음

|방법|설명|비고|
|--|--|--|
|tupleData = ()|튜플을 만들어 줌||
|len(tupleDate)|튜플의 항목 수를 반환||
|min(tuple), max(tuple)|튜플에서 가장 작은 값(min)과 가장 큰 값(max)을 반환||
|tupleData.count(value)|튜플에서 value의 개수를 반환||
|tupleData.index(value, position)|position 위치 이후에서 value가 있는 인덱스를 반환||


````python
def calculate_statistics(numbers):
  mean = sum(numbers) / len(numbers)
  variance = sum((x-mean)**2 for x in numbers) / len(numbers)
  return mean, variance

# 튜플로 반환된 값을 함수 인수로 사용
numbers = [1, 2, 3, 4, 5]
mean, variance = calculate_statistics(numbers)
print(f"Mean : {mean}, Variance: {variance}")
````

````python
def print_statistics(statistics):
  mean, variance = statistics
  print(f"Mean: {mean}, Variance:{variance}")

# 튜플을 인수로 전달
print_statistics((mean, variance))
````

## 딕셔너리(Dictionary)
- 중괄호 {}를 이용해서 딕셔너리 생성. 키(key)와 값(value)의 쌍으로 구성된 자료 구조
- 키는 유일(unique)해야 하며, 키에 리스트를 사용할 수 없음. but, 튜플 타입은 사용 가능
- 값은 중복이 가능하며, 모든 타입이 가능
- 키 목록에 없는 데이터를 사용하여 참조하면 에러가 발생, 인덱스를 이용한 참조는 지원안함
- 관계형 데이터를 다룰 때 특히 유용

|방법|설명|비고|
|--|--|--|
|dictData = {"key":"value", ...}|딕셔너리 생성||
|len(dictData)|딕셔너리 항목의 수를 반환||
|dictData.items()|딕셔너리의 각 항목들을(key, value) 형식의 튜플들로 반환||
|dictData.keys()|딕셔너리의 키(key)들을 반환||
|dictData.values()|딕셔너리의 value들을 반환||

````python
# 새로운 키-값 쌍 추가(insert) : 키를 지정하고 값을 추가로 입력
my_dict = {'apple':1, 'banana': 2}
my_dict['cherry'] = 3
print(my_dict)
````

- 저장한 순서대로 각 요소들이 저장되지 않으며, 중복을 허용 안함.
- True와 1은 같은 값으로 간주되며, False와 0도 같은 값으로 간주됨

### 집합 crud
- add()
  + 단일 요소를 집합에 추가할 때 사용
  + 기본 자료형과 튜플 추가함.
  + 리스트와 딕셔너리는 add()함수로 추가할 수 없음
  + 한 번에 하나의 요소만 추가 가능
  + 원본 집합을 직접 수정(반환값 없음.)
  ```
  my_set.add(element)
  ```
- update()
  + 기존 집합에 새로운 요소들을 추가할 때 사용
  + 리스트, 튜플, 딕셔너리, 문자를 추가함.
  + 딕셔너리는 키만 추가됨.
  + 여러 요소를 한번에 추가.
  + 중복된 요소는 자동으로 무시
  + 원본 집합을 직접 수정(반환값 없음.)
  ```
  existing_set.update(iterable)
  ```
- remove()
  + 특정 요소를 제거할 때 사용
  + 요소를 삭제함.
  + 함수의 인수는 셋에서 삭제할 값을 입력
  + 지정한 요소가 집합에 있으면 제거
  + 지정한 요소가 집합에 없으면 KeyError를 발생
  + 원본 집합을 직접 수정(반환값 없음.)
  ```
  my_set.remove(element)
  ```

## mutable(가변) vs immutable(불변)
- 가변 자료구조(mutable)
  + 생성 후 내용을 변경할 수 있음
  + 메모리 주소가 변경되지 않음
  + ex) 리스트(list), 딕셔너리(dictionary), 집합(set)
- 분변 자료구조 (immutable)
  + 생성 후 내용을 변경할 수 없음
  + 내용 변경 시 새로운 객체가 생성
  + ex) 튜플(tuple), 문자열(str), 정수(int), 실수(float)

````python
# 얕은 복사와 깊은 복사에서 변경 가능
# 데이터 타입의 복사 방식에 따라 결과가 달라질 수 있다
import copy

# 얕은 복사 (shallow copy)
list_a = [[1, 2], [3, 4]]
list_b = list_a.copy()
list_b[0][0] = 9
print(list_a)
print(list_b)
````

````python
# 깊은 복사 (deep copy)
list_c = [[1, 2], [3, 4]]
list_d = copy.deepcopy(list_c)
list_d[0][0] = 9
print(list_c)
print(list_d)

# copy()는 shallow copy 를 생성
# 즉, 새로운 객체를 만들지만 내부 객체에 대한 참조는 그대로 유지

# deepcopy()는 deep copy 를 생성
# 즉, 새로운 객체를 만들고 내부 객체도 모두 새로 생
````

## iterable (반복가능)
- 파이썬에서 순회할 수 있는 객체
- 요소를 하나씩 반환할 수 있는 객체
- for 루프에서 사용할 수 있음
- 리스트(list), 튜플(tuple), 문자열(str), 딕셔너리(dictionary), 집합(set), range 객체

````python
# 리스트를 이용한 예시
my_list = [1, 2, 3, 4, 5]
for item in my_list:
  print(item, end='')
print()
````

````python
# 문자열을 이용한 예시
my_string = "Hello"
for char in my_string:
  print(char, end='')
print()
````

````python
# range 객체를 이용한 예시
for num in range(5):
  print(num, end='')
print()
````

````python
# 딕셔너리를 이용한 예시
my_dict = {'a':1, 'b':2, 'c':3}
for key in my_dict:
  print(f"{key}:{my_dict[key]}", end='')
print()
````

````python
# iterable 객체를 list()로 변환
print(list("Python"))
````

## enumerate ()
- 반복 가능한 객체(iterable)를 인덱스와 함께 튜플 형태로 반환
- 반복문을 사용할 때 인덱스를 쉽게 접근 가능
- range() 함수를 이용해서 반복문의 인덱스를 만든 후 이를 이용해서 인덱스와 리스트 데이터를 처리하는 것은 직관적이지 못하고 코드가 어려워짐.
- 반복자(iterator) 또는 순서(sequence) 객체를 인수로 받음
- enumerate(iter)는 iter 객체를 (0, iter[0]), (1, iter[1]), (2, iter[2]), ... 형식으로 반환

````python
fruits = ['watermelon', 'orange', 'mango', 'grape', 'banana', 'apple']
for fruit in fruits:
  print("데이터는 {}입니다.".format(fruit))
````

````python
for i in range(len(fruits)):
  print("{}번째 데이터는 {}입니다.".format(i+1, fruits[i]))
````

````python
fruits = ['watermelon', 'orange', 'mango', 'grape', 'banana', 'apple']
for index, value in enumerate(fruits):
  print("{}번째 데이터는 {}입니다.".format(index+1, value))
````

````python
data = ['cat', 'dog', 'bird', 'fist']

for idx, value in enumerate(data):
  print(f"Index: {idx}, Value: {value}")
````

````python
# 가상의 학습 손실 값
losses = [0.9, 0.7, 0.5, 0.3]

# enumerate()를 사용하여 에포크와 손실 값을 출력
for epoch, loss in enumerate(losses):
  print(f"Epoch: {epoch}, Loss: {loss}")
````

````python
fruits = ['apple', 'banana', 'cherry']

for index, fruit in enumerate(fruits, start=1):
  print(f"Index: {index}, Fruit: {fruit}")
````

````python
# list comprehension 와 함께 사용
fruits = ['apple', 'banana', 'cherry']
indexed_fruits = [(index, fruit) for index, fruit in enumerate(fruits)]

indexed_fruits
````

````python
# 딕셔너리로 변환
fruits = ['apple', 'banana', 'cherry']
fruit_dict = {index: fruit for index, fruit in enumerate(fruits)}
fruit_dict
````

````python
# 문자열에서 문자와 인덱스 출력
text = "hello"

for index, char in enumerate(text):
  print(f"Index: {index}, Character: {char}")
````

````python
# 문자열 형태의 특성을 숫자로 인코딩할 때 enumerate()를 사용하여 각 문자에 고유한 숫자 인덱스를 부여
from sklearn.preprocessing import LabelEncoder

labels = ['apple', 'banana', 'cherry']
encoder = LabelEncoder()
encoded_laberls = encoder.fit_transform(labels)

print(list(enumerate(encoder.classes_)))
````

````python
# 모델 평가 시 enumerate()를 사용하여 각 샘플의 인덱스와 예측 결과를 매칭할 수 있음.
# 이를 통해 모델의 성능을 샘플 단위로 분석하거나, 오분류된 샘플을 확인 가능

from sklearn.metrics import accuracy_score

y_true = [0, 1, 0, 1, 0]
y_pred = [0, 0, 1, 2, 0]

for idx, (true, pred) in enumerate(zip(y_true, y_pred)):
  print(f"Sample {idx}: True={true}, Pred={pred}")

print(f"Overall accuracy: {accuracy_score(y_true, y_pred):.2f}")
````
