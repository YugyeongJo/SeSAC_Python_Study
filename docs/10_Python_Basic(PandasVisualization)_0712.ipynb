# 10. 판다스 시각화 (Pandas Visualization)

- 데이터를 시각적으로 표현하여 정보를 전달하고 이해하기 쉽게 만드는 과정
- 그래프, 차트, 지도, 다이어그램 등을 사용
- 데이터의 패턴, 추세, 이상치를 더 쉽게 파악 가능

## 데이터 시각화의 목적
- 데이터 이해
  + 데이터의 분포, 추세, 관계 등을 시각적으로 표현
  + 데이터의 특징과 패턴을 더 명확하게 볼 수 있음
- 통찰력 도출
  + 데이터를 분석하여 중요한 통찰력과 의사 결정을 도출
  + 복잡한 데이터를 시각적으로 단순화하여 중요한 정보를 강조
- 소통과 전달
  + 다른 사람과의 소통을 원활하게 하고, 데이터를 효과적으로 전달
  + 프레젠테이션, 보고서 등에서 데이터를 명확하게 제시

## 주요 그래프 유형
|그래프 유형| 설명| 예시|
|--|--|--|
|선 그래프|시간 변화에 따른 데이터를 시각화할 때 사용|주가 변동, 기온 변화 등|
|막대 그래프|범주형 데이터를 비교할 때 사용| 월별 매출, 인구 통계 등|
|히스토그램|데이터의 분포를 나타내며, 특정 구간에 속하는 데이터의 개수를 막대로 표현|시험 점수 분포, 연령 분포 등|
|산점도|두 변수 간의 관계를 나타내는 데 사용|키와 몸무게의 관계, 광고 비용과 매출 간의 관계|
|상자 그림|데이터의 분포와 이상치를 시각화하는 데 사용| 시험 점수의 분포, 연봉 분포 등|
|파이 차트|전체에 대한 각 부분의 비율을 나타냄| 시장 점유율, 예산 분배 등|
|히트맵| 데이터의 값을 색상으로 표현하여, 값을 쉽게 비교|상관 행렬, 지리적 데이터 등|


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 예제 데이터 생성
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.rand(100),
    'B': np.random.rand(100),
    'C': np.random.rand(100),
    'D': np.random.rand(100) * 100
})

# 선 그래프 (line)
plt.figure()
plt.plot(data['A'])
plt.title('Line Graph')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# 히스토그램 (hist)
plt.figure()
plt.hist(data['A'], bins=10)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 산점도 (scatter)
plt.figure()
plt.scatter(data['A'], data['B'])
plt.title('Scatter Plot')
plt.xlabel('A')
plt.ylabel('B')
plt.show()

# 박스 플롯 (box)
plt.figure()
plt.boxplot([data['A'], data['B'], data['C']])
plt.title('Box Plot')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()

# Seaborn 히트맵 (heatmap)
corr = data.corr()
plt.figure()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
````

## Matplotlib

- 장점
  + 다양한 그래프 유형을 지원
    * 선 그래프
    * 막대 그래프
    * 히스토그램
    * 산점도
    * 상자 그림
    * 3D 플롯
  + 매우 유연하고 세부적인 커스터마이징이 가능
    * 그래프의 모든 요소에 대해 세밀한 조정이 가능
    * 축의 눈금, 레이블, 제목, 범례 등을 사용자가 직접 설정
    * 그래프의 색상, 선 스타일, 마커 스타일 등도 사용자가 원하는 대로 변경 가능
  + 저수준 API를 제공하여 그래프의 모든 요소를 직접 제어
  + 다른 데이터 분석 라이브러리(Numpy, Pandas, SciPy 등)와의 통합이 잘 되어 있음
  + 다양한 출력 포맷 지원
    * 그래프를 여러 가지 파일 포맷(PNG, JPEG, PDF 등)으로 저장 가능
    * LaTeX와의 통합을 지원하여 과학적 논문 작성에 유용하게 사용

- 단점
  + 기본 스타일이 단조로움
  + 복잡한 시각화를 만들기 위해서는 코드의 길이가 길어질 수 있음

````python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 선 그래프 그리기
plt.plot(x, y, label='sin(x)')

# 그래프 제목, 축 레이블 추가
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 범례 추가
plt.legend()

# 그래프 보여주기
plt.show()

# 0부터 10까지 범위에서 사인 함수의 값을 계산하고, 선그래프로 표현
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# 예제 데이터셋 불러오기
tips = sns.load_dataset("tips")

# 스타일 설정
sns.set_style("whitegrid")
sns.set_palette("pastel")
sns.set_context("talk")

# 히트맵
plt.figure(figsize=(10, 8))
# 숫자 열만 선택하여 상관행렬 계산
numeric_cols = tips.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Tips Dataset")
plt.show()

# 페어플롯
sns.pairplot(tips, hue="sex")
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()

# 조인트플롯
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
plt.suptitle("Jointplot of Total Bill and Tip", y=1.02)
plt.show()

# 바이올린 플롯
plt.figure(figsize=(8, 6))
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True)
plt.title("Violin Plot of Total Bill by Day and Sex")
plt.show()

# 카운트 플롯
plt.figure(figsize=(8, 6))
sns.countplot(x="day", data=tips, palette="muted")
plt.title("Count Plot of Tips by Day")
plt.show()

# 스트립 플롯
plt.figure(figsize=(8, 6))
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title("Strip Plot of Total Bill by Day")
plt.show()

# 산점도
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, alpha=0.7)
plt.title("Scatter Plot of Total Bill and Tip by Sex")
plt.show()

# 박스플롯
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Box Plot of Total Bill by Day and Sex")
plt.show()

# 선형 회귀 플롯
plt.figure(figsize=(8, 6))
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Linear Regression Plot of Total Bill and Tip by Sex")
plt.show()
````
